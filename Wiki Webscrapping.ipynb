## Web scraping and analysis of list of largest companies by revenue from wikipedia

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from collections import deque
from bs4 import BeautifulSoup
import requests

```

```python
# Define the target URL
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
# Fetch the webpage content
page = requests.get(url)
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(page.text, 'html')
```

```python
# Print the HTML content for initial inspection
print(soup)
```

```python
# Extract the first table
table = soup.find_all('table')[0]
print(table)
```

```python
# Extract table header titles, clean them, and remove unnecessary columns
world_title = table.find_all('th')
world_title = [title.text.strip() for title in world_title]
world_title
```

```python
del world_title[7:] # Remove columns after the 7th 
```

```python
world_title
```

```python
# Create an empty DataFrame to store the extracted columns
df = pd.DataFrame(columns = world_title)
df
```

```python
# Extract all table rows
column_data = table.find_all('tr')
```

```python
# Iterate through rows (skipping the header row) and extract data
for row in column_data[2:]:
    row_data = row.find_all('td')
    individual_row_data = [row.text.strip() for row in row_data]
    dq = deque(individual_row_data) # Reorder columns
    dq.rotate(-7) # Shift columns 7 positions to the left
    individual_row_data= list(dq)
    del individual_row_data[7:] # Remove unnecessary data
    length = len(df)
    df.loc[length] = individual_row_data # Add the extracted row data to the DataFrame
    
```

```python
# Drop the 'Rank' column and rename remaining columns for clarity
df = df.drop('Rank', axis=1)
df = df.rename(columns={'Headquarters[note 1]' : 'Headquarters', 
                        'Revenue' : 'Revenue in Millions', 
                        'Profit':'Profit in Millions'}) 
```

```python
df
```

```python
# Get information about the DataFrame structure
df.info()
```

## Data Cleaning

```python
# Function to remove dollar signs from numeric columns
def remove_dollar_sign(x):
    return x.str.strip('$,')
```

```python
# Clean and convert 'Revenue in Millions' and 'Profit in Millions' columns to numeric format
df['Revenue in Millions'] = remove_dollar_sign(df['Revenue in Millions'])
df['Profit in Millions'] = remove_dollar_sign(df['Profit in Millions'])

df['Revenue in Millions'] = df['Revenue in Millions'].str.replace(',', '', regex=False).astype(int)
df['Profit in Millions'] = df['Profit in Millions'].str.replace(',', '', regex=False).astype(int)
```

```python
# Clean and convert 'Employees' column to numeric format 
df['Employees'] = df['Employees'].str.replace(',', '', regex=False).astype(int)
```

```python
df
```

```python
# Get information about the DataFrame structure after cleaning
df.info()
```

```python
# Calculate total revenue by industry
revenue_by_industry = df.groupby('Industry')['Revenue in Millions'].sum().sort_values(ascending=False)
revenue_by_industry
```

```python
# Create a bar chart to visualize revenue by industry
revenue_by_industry.plot(kind='barh')
plt.xlabel('Revenue in USD millions')
plt.title('Total Revenue by Industry')
plt.show()
```

```python
# Find the top 3 countries with the most company headquarters
df.groupby('Headquarters')['Revenue in Millions'].sum().head(3)
```

```python
# find countries with the most headquarters
country_with_the_most_hq = df.groupby('Headquarters')['Headquarters'].count().sort_values(ascending=False)
country_with_the_most_hq
```

```python
# Create a bar chart to visualize countries with the most headquarters
country_with_the_most_hq.plot(kind='bar')
plt.title('Countries with the most headquarters')
plt.ylabel('')
plt.show()
```

```python
# Find the top 3 industries with the most employees
industry_with_the_most_employee = df.groupby('Industry')['Employees'].sum().head(3)
industry_with_the_most_employee
```

```python
# Create a pie chart to visualize industry with the largest employee count
plt.figure(figsize=(10,6))
explode = (0, 0.3, 0.0)
industry_with_the_most_employee.plot(kind='pie', explode=explode, autopct = '%1.1f%%')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.title('Pie chart of industry with the largest employee')
plt.show()
```
